# Preparing a new python environment

If you want to replicate the output of these notebooks exactly, it's strongly suggested you create an environment with specific versions of libraries used in this tutorial. There are a few ways to do this:

### Using conda or [mamba](https://mamba.readthedocs.io/en/latest/installation/mamba-installation.html)

If you use conda (or mamba, they are pretty much interchangeable), you can use the `env.yaml` file in this repo to install all the required packages as simple as:

```bash
mamba env create -n parlacaptutorials -f env.yaml
```
and then selecting `parlacaptutorials` for your notebook interpreter.

If you'd like to start using mamba, check out [the documentation](https://mamba.readthedocs.io/en/latest/installation/mamba-installation.html).

### Using venv

First, assure you have python 3.14.2 installed. Then create a new environment with
```bash
python3.14 -m venv parlacaptutorials
```
and activate it by running

```
source parlacaptutorials/bin/activate 
```

If you happen to use a Windows-flavoured operating system, the above command won't work, substitute it with 

```powershell
parlacaptutorials\Scripts\activate
```

Finally, install the required libraries:

```bash
pip install -r requirements.txt
```
