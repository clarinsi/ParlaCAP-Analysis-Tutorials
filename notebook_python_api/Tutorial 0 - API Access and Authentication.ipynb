## Preparing a new Python environment

If you want to replicate the output of these notebooks exactly, it is strongly recommended to create a fresh environment with the library versions used in this tutorial.

There are two easy ways to do this.

### Using conda or mamba

If you use conda or mamba, you can create the environment from the `env.yaml` file in this repository:

```bash
mamba env create -n parlacaptutorials -f env.yaml
```

Then select `parlacaptutorials` as your notebook interpreter.

### Using venv

First, make sure you have Python 3.14.2 installed.

Create a new environment with:

```bash
python3.14 -m venv parlacaptutorials
```

Activate it with:

```bash
source parlacaptutorials/bin/activate
```

If you use Windows, use this instead:

```powershell
parlacaptutorials\Scripts\activate
```

Then install the required libraries:

```bash
pip install -r requirements.txt
```



# Obtaining an API key

An API (Application Programming Interface) is a structured way for your code to request data from an external system. In this case, instead of manually downloading the entire parliamentary speech dataset via a website, you send a request to a server through the API and receive back only the specific data you need (e.g. filtered by country, time period or topic). This makes working with large datasets much easier and it keeps the data up to date. To use an API, you typically need to authenticate first, which is what Tutorial 0 walks you through before you start the actual analyses.

You can use the cell below to save the API key to disk:

We can test if the key works with the cell below:

````python
import requests

# Read the key from my_secrets.env.
headers = {"X-API-Key": open("parlacapi_key").read()}

# Ask the API for one row.
query = {
    "limit": 1,
    "offset": 0,
    "columns": ["date", "cap_category", "parliament"],
}

response = requests.post(
    "https://parlacap.ipipan.waw.pl/api/v1/filter", json=query, headers=headers
)
response.raise_for_status()
response.json()["rows"][0]
````
